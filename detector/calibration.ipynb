```python
def func_to_fit2(x, a, b, c, d):
    return -a*np.tan((x + b)/c) + d


fit_grid = np.linspace(min(data12345_V)-3, max(data12345_V)+3, 1000)

doFit = False
if doFit:
    popt, pcov = curve_fit(func_to_fit2, data12345_V, data12345_x)    
    fit = func_to_fit2(fit_grid, *popt)
else:
    fit = func_to_fit2(fit_grid, 1, -4, 3, 25)

plt.figure()
plt.plot(data12345_V, data12345_x)
plt.plot(fit_grid, fit)
plt.ylim(0,50)
plt.xlim(-3, 3)
plt.show()
```

```python
data_V = data3_V
data_x = data3_x

def func_to_fit3(x, a, b, c):
    return a/(x+b) + c
    #return a*np.exp(x + b) + c

popt, pcov = curve_fit(func_to_fit3, data_V, data_x)

fit_grid = np.linspace(min(data_V), max(data_V), 1000)
fit = func_to_fit3(fit_grid, *popt)

fit_grid = np.linspace(min(data_V), max(data_V), 1000)
fit = np.polynomial.polynomial.Polynomial.fit(data_V, data_x, 5)

plt.figure()
plt.plot(data_V, data_x)
plt.plot(fit_grid, fit(fit_grid))
plt.show()

print(fit.convert().coef)
```
